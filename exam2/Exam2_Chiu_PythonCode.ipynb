```python
# for small samples (<50) we use t-statistics
xbar = 19.5 # sample mean
sd = 5.2
n = 65 # sample size
df = n-1 # degrees of freedom

# define T-score function
def Tscore(ci, n):
  t = stats.t.ppf(1- ((100-ci)/2/100), n-1) # CI, df
  return t

# print(Tscore(90, 65))

# 90% CI
ci = 90
tscore = Tscore(ci, n)
lower = xbar - tscore * sd/math.sqrt(n) #lower level
upper = xbar + tscore * sd/math.sqrt(n) # upper level
print("The population mean is {} to {} with {}% confidence intervals".format(lower, upper, ci))

# 95% CI
ci = 95
tscore = Tscore(ci, n)
lower = xbar - tscore * sd/math.sqrt(n) #lower level
upper = xbar + tscore * sd/math.sqrt(n) # upper level
print("The population mean is {} to {} with {}% confidence intervals".format(lower, upper, ci)) 
```
