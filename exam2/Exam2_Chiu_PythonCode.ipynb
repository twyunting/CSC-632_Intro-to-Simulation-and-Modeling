Repeat Part 3a) using a Python code. Provide the code file in a .py file and provide comments within the code. For this problem in python use a t-test (as described in the class). 


```python
# for small samples (<50) we use t-statistics
xbar = 19.5 # sample mean
sd = 5.2
n = 65 # sample size
df = n-1 # degrees of freedom

# define T-score function
def Tscore(ci, n):
  t = stats.t.ppf(1- ((100-ci)/2/100), n-1) # CI, df
  return t

# print(Tscore(90, 65))

# 90% CI
ci = 90
tscore = Tscore(ci, n)
lower = xbar - tscore * sd/math.sqrt(n) #lower level
upper = xbar + tscore * sd/math.sqrt(n) # upper level
print("The population mean is {} to {} with {}% confidence intervals".format(lower, upper, ci))

# 95% CI
ci = 95
tscore = Tscore(ci, n)
lower = xbar - tscore * sd/math.sqrt(n) #lower level
upper = xbar + tscore * sd/math.sqrt(n) # upper level
print("The population mean is {} to {} with {}% confidence intervals".format(lower, upper, ci)) 
```

# 6d.
Write a Python code to determine and address the hypothesis tests that you developed in Part 6a) and manually answered in Part b). 

```python
# Ho: workers = 0.113 vs Ha: workers > 0.113.
n = 400 
x = 52
p = 0.113
alpha = 0.05

# compute the sample proportion
phat = x/n
# print(phat)

# compute the z-score
z = (phat-p) / np.sqrt(p*(1-p)/n)
# print(z)

# find the p-value
# if two-sided, needs to *2, otherwise is not
pvalue = stats.norm.sf(abs(z))
print(" The p-value is",pvalue)

# state the hypothesis test
if pvalue < alpha:
  print("We have 95 % confident to reject the null hypothesis")
else:
  print("We are 95 % confident fail to reject the null hypothesis")
```
